```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.thehindu.com/news/national/other-states/pramod-sawant-meets-pm-modi-in-delhi/article65231138.ece?homepage=true"
r = requests.get(url)		# r variable has all the HTML code
htmlContent = r.content	# r returns response so if we want the code we write r.content
# print(htmlContent)	
```

```python
soup = BeautifulSoup(r.content, 'html.parser')

# print(soup.prettify())
```

```python
s1=""
for i in soup.find_all('p'):
    a=(i.text)
    s1+=(i.text)
    # print(a) 
# s1.strip()  
while "\n" in s1:
    s1=s1.replace("\n"," ")
while "     " in s1:
    s1=s1.replace("     "," ")
while "    " in s1:
    s1=s1.replace("    "," ")    
# s1.strip()    
# print(s1)
```

```python
import re
import requests
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

punc='''!()-[]{};:'"\,<>./?@#$%^&*_~”’“‘|—'''
stop_word=["a","the","an","is"]
for ele in s1:
    if ele in punc:
        s1 = s1.replace(ele, "")
no_of_words=list(s1.split())
print(len(no_of_words)) 
s2=""
for i in no_of_words:
    s2+=i
    s2+=" "
print(s2)
stop_words = stopwords.words('english')
l=list(s2.split())
for x in range(len(l)):
    if l[x] in stop_words:
        l[x]=""
s_inf=""
for x in l:
#     p=''.join([i for i in x if not i.isdigit()])
#     p=re.sub(r'[0-9]+', '', x)
#     if x.isnumeric():
#         continue
    s_inf+=x
    s_inf+=" "
# print(s_inf)      
num=['1','2','3','4','5','6','7','8','9','2014']
for ele in s_inf:
    if ele in num:
        s_inf= s_inf.replace(ele,"")
print(s_inf)        
# result = ''.join([i for i in s_inf if not i.isdigit()])
# print(result)
```

```python
di={}
for x in range(len(l)):
    b=l[x].lower()
    if b=="" or b.isnumeric():
        continue
    a=l.count(b)
    di[b]=a
print(di)    
```

```python
def get_value(tp):
    return di[tp]
# l=sorted(di,key=get_value, reverse=True) 
l=[]
i=1
for x in sorted(di,key=get_value, reverse=True):
    ll=[]
    ll.append(i)
    ll.append(x)
    ll.append(di[x])
    l.append(ll)
    i+=1
#     print(x,di[x])
# print(l)
print(l)
```
